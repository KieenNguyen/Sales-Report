<h1> II.Calculate Conversion Rate

<h4> Conversion_rate =  Number of Buyers / number of visitors

<H4> To solve this problem need to have number of buyers and number of visitor each day 

```python
#import data:
cust_info = pd.read_csv('visitor_information.csv')
cust_info
```

<H5> 1.Find number of visitor each day:

*Company record the visitor at the first time they come visit company site, store,... So all information in cust_info sheet is considered to be first time visit*

```python
#Find number of visitor every day of the company
visit_every_day = cust_info.groupby(['create_date','visitor_source']).count().reset_index().rename(columns = {"visitor_id":"number visitor"})
#Drop unnessary columns:
visit_every_day = visit_every_day.drop(columns = ['gender','age'])
visit_every_day
```

```python
visit_every_day['create_date'] = pd.to_datetime(visit_every_day['create_date'])
```

<H5> 2.Find number of new buyer order every day:

*Number of buyers at the first time visit company is calculate by count number of buyers who have date visit = day buy*

```python
#Create Dataframe contents first time buy of all customers company currently have: 
#To have first buy sort order day and keep the first day: 
data_sort = data_sort.sort_values(by='date')
first_time = data_sort.drop_duplicates(subset= "visitor_id", keep="first")
first_time
```

<H5> Find customer who first come and first buy: 

```python
#Merge cust_info and first_time buyers sheet together:
first_visitbuy = pd.merge(cust_info, first_time,how ='inner',left_on=['visitor_id'], right_on=['visitor_id'])
first_visitbuy
```

```python
#Convert create_date data type into datetime: 
first_visitbuy['create_date']=pd.to_datetime(first_visitbuy['create_date'])
```

```python
first_visitbuy.info()
```

```python
#Loc ID that have first buy = with fist visit: 
first_visitbuy =first_visitbuy.loc[first_visitbuy['create_date'] == first_visitbuy['date']]
first_visitbuy
```

```python
#Count number of ID first visit and first buy: 
first_visitbuy = first_visitbuy.groupby(['create_date','visitor_source']).count().reset_index()
first_visitbuy = first_visitbuy.drop(columns = ['Day','Month','Year','old_cust','age','date','gender','no_of_order'])
first_visitbuy
```

```python
#Rename columns: 
first_visitbuy = first_visitbuy.rename(columns ={'visitor_id':'first_visitbuy'})
first_visitbuy
```

```python
#create convertion rate Data frame: 
conversion_rate = pd.merge(first_visitbuy, visit_every_day, how= "outer", on=['create_date','visitor_source'])
conversion_rate
```

```python
# NA appears because there will be days with visits but no buyers,
#to solve it, it is necessary to add 0 to the NA positions
conversion_rate['first_visitbuy'] = conversion_rate['first_visitbuy'].fillna(0)
conversion_rate
```
