```python
#import library:
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

<H1>DATA PREPARATION

```python
#Check data in sale data file: 
sale_raw= pd.read_excel('data.xlsx')
sale_raw.head()
```

```python
#Sale data need to be unpivoted to discover more: 
sale= sale_raw.melt(id_vars =['visitor_id'], var_name ='date', value_name ='no_of_order')
sale
```

```python
#Drop day which have no order occurred:
sale_data= sale.drop(labels=sale.loc[sale['no_of_order']==0].index)
sale_data
```

```python
#43770, 43771,...,43896 each number presents for each days from 1.11.2019
#Now I need to covert those number into datetime, excel handle this task better than Python, so I will export 
#this data set and covert in excel to save time: 
#save data: 
sale_data.to_excel("Sales data unpivot.xlsx", index = False)
```

```python
#Import data with correct date time: 
sale_final= pd.read_excel('Sales_data_unpivot_correct datetime.xlsx')
sale_final
```

```python
sale_final.info()
```

```python
#Create Month,Day, Year columns for furthur analysis:
sale_final['Day'] = pd.DatetimeIndex(sale_final['date']).day
sale_final['Month'] = pd.DatetimeIndex(sale_final['date']).month
sale_final['Year']= pd.DatetimeIndex(sale_final['date']).year
sale_final
```

```python
#Export Data to answer question 2:
sale_final.to_excel("Sale Data Final.xlsx", index = False)
```
